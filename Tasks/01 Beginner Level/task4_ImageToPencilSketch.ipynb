 ### 01 Beginner Level Task - Image to Pencil Sketch with Python



Steps:
> 1. Importing the dependencies
2. Importing and reading the original image
3. Converting the original Image to a grayscale image
4. Checking the maximum and minimum values of the pixels in the original image
5. Inverting the image
6. Blurring the image
7. Pencil Sketch image
8. Saving the image to a local directory

#### Importing the dependencies

```python
# Importing the required libraries
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Importing and reading the original image

```python
# Reading the Kingfisher image from the local directory
img = cv.imread('./Data/Kingfisher.jfif')
plt.figure(figsize=(7, 7))
plt.title('Original Image')
plt.imshow(img)
```

#### Converting the original Image to a grayscale image

```python
# Converting the original image to a grayscale image by setting vmin value to 0 and cmap to 'gray'
gray_img = cv.imread('./Data/Kingfisher.jfif', 0)
plt.figure(figsize=(7, 7))
plt.title('Grayscale Image')
plt.imshow(gray_img, cmap='gray')
```

#### Checking the maximum and minimum values of the pixels in the original image

```python
# Max and min values of the pixels
pixels = {
    'Minimum': img.min(),
    'Maximum': img.max()
}
print(list(pixels.keys())[0], "value of the pixels present in the image:", pixels['Minimum'])
print(list(pixels.keys())[1], "value of the pixels present in the image:", pixels['Maximum'])
```

#### Inverting the image

```python
# Inverting the grayscale image by subtracting 255 from each pixel of the grayscale image (Inversion)
inv_img = 255 - gray_img
plt.figure(figsize=(7, 7))
plt.title('Inverted Image')
plt.imshow(inv_img, cmap='gray')
```

#### Blurring the image

```python
# Blurring the image using Gaussian Blur to eliminate all the detail imformation and noise present in the image
blur_img = cv.GaussianBlur(inv_img, ksize=(21, 21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(7, 7))
plt.title('Blurred Image')
plt.imshow(blur_img, cmap='gray')
```

#### Pencil Sketch image

```python
# Applying Dodging and Burning technique to get the final Pencil Sketch like looking image
dodge_burn = lambda image, mask: cv.divide(image, 255 - mask, scale=250)
pencil_sketch = dodge_burn(gray_img, blur_img)
plt.figure(figsize=(7, 7))
plt.title('Pencil Sketch Image')
plt.imshow(pencil_sketch, cmap='gray')
```

#### Saving the image to a local directory

```python
# Pass the directory path and the final image to the function.
# Successful creation of the image will return True.
cv.imwrite('./Data/KingfisherPencilSketch.jpeg', pencil_sketch)
```
